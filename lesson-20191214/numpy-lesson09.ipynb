```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. 
# If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, 
# and automatically generates the x values for you. Since python ranges start with 0, the default x vector
# has the same length as y but starts with 0. Hence the x data are [0,1,2,3].

# 默认认为是 Y (如果是 单个数组的话)

plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo')      # yo  yellow  circle marker >>>> 'o'	circle marker
# The axis() command in the example above takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes.
plt.axis([0, 6, 0, 20])    
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
 
x = np.arange(1, 11) 
y =  2  * x +  5 
plt.title("First graph") 
 
# fontproperties 设置中文显示，fontsize 设置字体大小
plt.xlabel("x")

plt.ylabel("y")

plt.plot(x,y) 

plt.show()
```
